## Exploratory Data Analysis.

### Question 1 Which industry has the most number of fraudulent jobs?

```python
fraud_count =df.groupby('industry').sum().sort_values(by='fraudulent',ascending=False).head(10)
```

```python
fraud_count
```

```python
plt.figure(figsize=(11,9))
fraud_count['fraudulent'].plot(kind='bar')
plt.ylabel('Number of fake job postings')
plt.title('Top ten industries with highest fake job postings');
```

### Conclusion

> 1. The industry with the highest number of fake job postings is the Oil & Energy industry with 109 fake job postings
